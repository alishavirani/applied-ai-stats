```python
import numpy as np
import pylab
import scipy.stats as stats

#N(0,1) standard normal dist
#loc = mean, scale=std
#size=1000 says that generate random nos (sample randomly) from a GD with mean=0,sd=1 and generate 1000 such samples, ie 1000 samples b/w -1 and 1 which are GD
std_normal = np.random.normal(loc = 0, scale = 1, size =1000)

#0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal, i)) #computes percentiles
```

```python
#generate 100 samples from N(20, 5)
#mean=20, sd=5, 100 samples
measurements = np.random.normal(loc=20, scale=5, size=100)
#try size=1000

#probplot or qq plot in scipy library
#qq plot where measurements is the distribution which we want to check if it is GD or not, dist is the parameter to check for which dist we want to compare measurements with, norm means compare with standard normal variate ie N(0,1)
#plotting using pylab lib
#x-axis is N(0,1)
#y-axis is N(20,5)

#since both the x-axis and y-axis follow the GD, ie X and Y RV are normal, then mostly all of the points lie on the red line
#instead of taking 100 size, we make it 1000, more of the points fall on the straight line
#instead of taking 100 size, we make it 50, there is more deviation from the straight line
#as the no of sample size increases, more of the points lie on the line
#if the no. of observations/samples is small, it is hard to interpret QQ Plot. This is the limitation of QQ plot
stats.probplot(measurements, dist="norm",plot=pylab)
pylab.show()
```

```python
#generate 100 samples from N(20, 5) from uniform dist
measurements = np.random.uniform(low=1, high=100, size=100)

#x-axis=GD, y-axis=uniform dist
#points do not lie on the line, points are going far away from the straight line
#In the extremities points are diverging to a great extent
#If we increase the size to 10000, then the points diverge more and more
stats.probplot(measurements, dist="norm",plot=pylab)
pylab.show()
```

```python
If most of the points lie on the straight line, then we can conclude that the distrbution on the x and y are of the same family.
Y axis can be any distribution, not just normal
```
